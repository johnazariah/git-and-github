# Exploratory Analysis — Climate Data

This notebook explores the historical temperature dataset before formal analysis.

## Goals
- Understand the shape and quality of the data
- Identify missing values and outliers
- Generate initial visualisations

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Configure plots
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)
```

## Load and Inspect Data

```python
# Placeholder — replace with your data file path
# df = pd.read_csv('data/temperatures.csv', parse_dates=['date'])

# For now, generate sample data
np.random.seed(42)
dates = pd.date_range('2000-01-01', '2023-12-31', freq='D')
stations = ['Sydney', 'Melbourne', 'Brisbane']

records = []
for station in stations:
    base_temp = {'Sydney': 22, 'Melbourne': 20, 'Brisbane': 25}[station]
    temps = base_temp + 5 * np.sin(2 * np.pi * np.arange(len(dates)) / 365) + np.random.normal(0, 2, len(dates))
    # Add a slight warming trend
    temps += np.linspace(0, 1.5, len(dates))
    for d, t in zip(dates, temps):
        records.append({'date': d, 'station': station, 'temperature': round(t, 1)})

df = pd.DataFrame(records)
print(f'Dataset: {len(df):,} records from {df["station"].nunique()} stations')
df.head()
```

## Summary Statistics

```python
df.groupby('station')['temperature'].describe().round(1)
```

## Temperature Distribution by Station

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5), sharey=True)
for ax, station in zip(axes, stations):
    data = df[df['station'] == station]['temperature']
    ax.hist(data, bins=50, alpha=0.7, edgecolor='black')
    ax.set_title(station)
    ax.set_xlabel('Temperature (°C)')
axes[0].set_ylabel('Count')
plt.tight_layout()
plt.show()
```
