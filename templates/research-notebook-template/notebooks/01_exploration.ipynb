# 01 — Data Exploration

Initial exploration of the dataset. Goals:
- Understand shape and structure
- Check for missing values and outliers
- Generate summary statistics and initial plots

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Import shared functions from src/
from src.data import load_data, describe_dataset

plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)
```

## Load Data

```python
df = load_data()
describe_dataset(df)
```

## Summary Statistics

```python
df.describe()
```

## Distribution Plots

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5), sharey=True)
for ax, station in zip(axes, df['station'].unique()):
    subset = df[df['station'] == station]
    ax.hist(subset['temperature'], bins=50, alpha=0.7, edgecolor='black')
    ax.set_title(station)
    ax.set_xlabel('Temperature (°C)')
axes[0].set_ylabel('Count')
plt.tight_layout()
plt.show()
```
